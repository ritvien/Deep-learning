# Perceptron Model

known as a **`Linear Binary Classifier.`**

* **Perceptron** is a **single layer** neural network
* a **multi-layer** perceptron is called **Neural Networks**.

**Mô hình tổng quát**

* Mỗi mô hình luôn có <span style="color: red;"> **1 input layer**</span> , <span style="color: green;"> **1 Output layers**</span> , có thể có hoặc không <span style="color: blue;"> **Hidden layer**</span>.
* Perceptron là các mô hình chỉ có <span style="color: red;"> **1 input layer**</span> và <span style="color: green;"> **1 Output layers**</span>
* Tổng số layer trong mô hình quy ước là số layer - 1 (không tính input layer)
* Mỗi hình tròn là 1 node, được gọi là 1 unit

## How it work?

**Step 1:** All the inputs **x** are multiplied with their weights **w**

**Step 2**: Add all the multiplied values and call them **Weighted Sum**

<font size=5>$ \sum_{i=0}^n w_i.x_i + b$
    <font>

**Step 3**: Apply that weighted sum to the correct `Activation Function`.

## <span style="color: red;"> Activation function</span> 

* is a simple function that output 0 or 1
* usually use **Non-linear** Activation Function

Some terminologies (thuật ngữ):
* **Derivative** or **Differential**(Đạo hàm và Vi phân): Change in y-axis w.r.t. change in x-axis.It is also known as slope(độ dốc)

* **Monotonic function**(Hàm đơn điệu): A function which is either entirely non-increasing or non-decreasing.

### Linear Activation Function

A very simple way:

<font size=5> $f(x) = x$

<span style="color:red">&#10060;</span> **Problems:**

👉 Không thể sử dụng ``Backpropagation `` vì đạo hàm là 1 hằng số

👉 **Tất cả các layers** của mạng đều có thể chuyển về thành **1 hàm linear**. Do đó, không cần biết có bao nhiêu layers, cuối cùng cũng có thể biểu diễn đơn giản từ input đến output chỉ bằng 1 hàm linear

### Sigmoid Function

* Is **differentiable**, means we can find the slope at any two points
* Is **monotonic**, but the function's derivative is not

<span style="color: green; font-size: 24px;">&#10004;</span>The output of the sigmoid activation function can be interpreted as a probability

👉 can cause a neural network to get stuck at the training time.

### Tanh or hyperbolic tangent

<font size = 5> $ f(x) = \frac{e^x-e^{-x}}{e^x + e^{-x}} = \frac{2}{1+e^{-2x}}-1 $

* Is **differentiable**
* Is **monotonic**, but the function's derivative is not

<span style="color: green; font-size: 24px;">&#10004;</span> Using the tanh activation function results in **higher values of gradient** during training and higher updates in the weights of the network. **So, if we want strong gradients and big learning steps**, we should use the tanh activation function.

&#128161; Both **tanh** and **logistic sigmoid** activation functions are used in **`feed-forward nets.`**

**Feed-forward nets**: Mạng truyền thẳng

* Là một loại mạng nơ-ron nhân tạo (ANN)
* Thông tin di chuyển 1 chiều
* Từ lớp đầu vào qua các lớp ẩn và đi đến lớp đầu ra mà **không có bất kỳ kết nối phản hồi nào**
* Training bằng ``Backpropagation``

**Backpropagation**

là 1 phương pháp đào tạo mạng neural, đặc biệt là các deep neural net works

* Step 1: Đưa dữ liệu vào mạng và truyền qua mỗi lớp từ đầu vào đến đầu ra, tạo ra dự đoán
* Step 2: So sánh với giá trị thực tế để đo lỗi
* Step 3: Backward Pass: lỗi này được truyền ngược lại từ lớp đầu ra đến các lớp trước đó để cập nhật trọng số
* Step 4: Sử dụng GD để điều chỉnh trọng số, làm giảm thiểu lỗi bằng cách điều chỉnh trọng số
* Step 5: Lặp lại quá trình

### ReLU

Rectified Linear Unit : 

<font size =5> f(s) = max(0,s)

<img src='https://i.stack.imgur.com/YKS2g.png' style="width:500px;"/> 

Được sử dụng rộng rãi gần đây vì tính **đơn giản** của nó. Ưu điểm của nó:
* Tính toán đơn giản
* Tăng tốc độ rất nhanh, đạo hàm rất đơn giản
* Thể hiện giá trị giúp đơn giản hóa model nhờ giá trị âm được lưu trữ là 0 thay vì các giá trị gần 0 như các activation funtion khác
* Xử lý tuyến tính
* Hỗ trợ huấn luyện deep network

### Softmax

Đối với bài toán có rất **nhiều classes (multi-class)**, các binary classifiers mặc dù có thể áp dụng cho các bài toán multi-class, chúng vẫn có những hạn chế nhất định.

Khi đó nên sử dụng **Softmax**

👉<font size=5><span style="color: red;">$ x$ </span> $= [x_0,x_1,....x_d]$

👉<font size=5> <span style="color: green;">$W^T$</span>$ = [w_{c1_{0}},w_{c1_{1}},...,w_{c1_{d}},$     <span style="color: #03e3fc;"> --->Class1

<font size=5><div style="margin-left: 120px;">$w_{c2_{0}},w_{c2_{1}},...,w_{c2_{d}},$<span style="color:#03e3fc;">--->Class2 
    
<font size=5><div style="margin-left: 120px;">.... 
    
<font size=5><div style="margin-left: 120px;">$w_{cC_{0}},w_{cC_{1}},...,w_{cC_{d}}]$<span style="color:#03e3fc;">--->Class C

👉<font size=5> $z = $<span style="color: green;">$W^T$</span><span style="color: red;">$ x$ </span> $ =[z_1,z_2,...,z_C]$

<font size =6> &#x1F9EE; : $a_i = \frac{e^{z_i}}{\sum_{i=1}^C e^{z_i}} \in (0,1) $

<font size=6> $ a = [a_1,a_2,...,a_c]$ </font>tương ứng với xác suất của các lớp Output

# Feed-forward nets.

👉 Thực hiện tính toán từ đầu đến cuối của Network

👉 Tính *predicted output $\hat{y}$* với một input **x**:

<font size=5> $a^{(0)} = x$ : input


<font size=5> $z^{(l)} = W^{(l)T}a^{(l-1)} + b^{(l)}, l = 1,2,...L$

<font size=5> $a^{(l)} = f(z^{(l)})$, l = 1,2,...,L  với f là **activation funtion**

<font size=5> $ \hat{y} = a^{(L)} $: last $a^{(l)}$

# Backpropagation 

👉 Việc ta cần tối ưu là tìm $W,b$ để $J(W,b,X,Y)$ là một hàm mất mát đạt giá trị nhỏ nhất

<span style="color:red">&#10060;</span> Việc áp dụng các `gradient-based` để tối ưu dường như không thể thực hiện do các hàm mất mát không phụ thuộc trực tiếp vào các hệ số để thực hiện đạo hàm:

<font size=5><div style="margin-left: 200px;"> $J(W,b,X,y) = \sum_{n=1}^N||y_n-\hat{y}_n||_2^2$
    
<font size=5><div style="margin-left: 367px;">$ =\sum_{n=1}^N||y_n-a_n^{(L)}||_2^2$

*Với N là số cặp dữ liệu (x,y) trong tập training*

&#128161; `Backpropagation` :👉 Tính đạo hàm **ngược** từ layer cuối cùng đến layer đầu tiên.

<div style="margin-left: 150px;">👉 layer cuối đc tính trước vì nó gần gũi hơn với predicted outputs và hàm mất mát
    

<span style="color: green; font-size: 24px;">&#10004;</span> Dựa trên quy tắc <span style="color: blue;"> *chain rule*</span> , tính đạo hàm của hàm hợp

<font size=5> $J(W,b,X,y) =\sum_{n=1}^N||y_n-$ <span style="color: red;">$ a_n^{(L)}$</span>  $||_2^2$

 <font size=5>thay:<span style="color: red;">$ a_n^{(L)} = f($  <span style="color: #02c405">$z^{(L)}$</span>$)$</span>

<font size=5>👉 $J(W,b,X,y) =\sum_{n=1}^N||y_n-$ <span style="color: red;">$f($<span style="color:#02c405;">$z^{(L)}$</span>$)$</span>  $||_2^2$

👉 Để áp dụng cho GD, ta phải tính đạo hàm theo mỗi biến **W** và **b**  

<font size=5><span style="color: #02c405"> $z^{(L)} = W^{(L)T}a^{(L-1)} +b^{(L)} (*) $</span> 

Đạo hàm theo **W**: <font size=5> $\frac{\delta J(W,b,X,y)}{\delta W^{(L)}}$ <span style="color: blue;"> =</span>  $\frac{\delta J}{\delta z^{(L)}} $<span style="color: #02c405;">$\frac{\delta z^{(L)}}{\delta W^{(L)}}$ =</span> $\frac{\delta J}{\delta z^{(L)}}$<span style="color: #02c405;">$a^{(L-1)} $</span> = $e^{(L)}$<span style="color: #02c405;">$a^{(L-1)} $</span>

Đạo hàm theo **b**:<font size=5> $\frac{\delta J(W,b,X,y)}{\delta b^{(L)}}$ <span style="color: blue;"> =</span>  $\frac{\delta J}{\delta z^{(L)}} $<span style="color: #02c405;">$\frac{\delta z^{(L)}}{\delta b^{(L)}}$ =</span> $\frac{\delta J}{\delta z^{(L)}}$ = $e^{(L)}$

Trong đó:
* <span style="color: blue;"> =</span> là biến đổi <span style="color: blue;"> *chain rule*</span> 
* <span style="color: #02c405"> =</span> là biến đổi đạo hàm của <span style="color: #02c405">$(*)$

<span style="color: green; font-size: 24px;">&#10004;</span> Từ đó sẽ có đạo hàm của **J** theo **W** và **b** để áp dụng thuật toán tối ưu GD

# Keras Basics

```python
import keras
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**y = mx + b + noise**

```python
m,b = 2,3
x = np.linspace(0,50,100)
```

```python
np.random.seed(100)
noise = np.random.normal(loc=0,scale=4,size=len(x))
```

```python
y = 2*x + b + noise
```

```python
plt.plot(x,y,'*')
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(units=4,input_dim=1,activation='relu'))

model.add(Dense(4,activation='relu'))

model.add(Dense(1,activation='linear')) #output layer

model.compile(loss='mse',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=200)
```

```python
model.history.history['loss']
```

```python
loss = model.history.history['loss']
epochs = range(len(loss))
plt.plot(epochs,loss)
```

```python
x_for_predictions = np.linspace(0,50,100)

y_pred = model.predict(x_for_predictions)
y_pred
```

```python
plt.plot(x,y,'*')
plt.plot(x_for_predictions,y_pred,'r')
```
